```python
# Preprocess text
input_text = '\n'.join( list(df[5].astype(str)) );
```

```python
# Save text to file
with open('./data/input.txt', 'w', encoding='utf-8') as file:
    file.write(input_text)
```

```python
# Train model
model_name = 'tokenizer'
spm.SentencePieceTrainer.train(input='./data/input.txt', model_prefix=model_name, vocab_size=1000)
```

```python
# Load model
model = spm.SentencePieceProcessor()
model.load(f'{model_name}.model')
```

```python
# Tokenize text
text = input_text  # TODO: use a testing dataset
tokens = model.encode_as_pieces(text)
```
